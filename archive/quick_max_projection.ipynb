```python
from specpy import *
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [10,10]
```

```python
im = Imspector()
ms = im.active_measurement()
stack_names = ms.active_configuration().stack_names()

for stack_name in stack_names:
    img = ms.active_configuration().stack(stack_name).data()
    img = img.squeeze()
    plt.figure()
    plt.imshow(np.apply_along_axis(np.max, 0, img))
    plt.show()
```
