```python
from specpy import *
import sys, time, os
```

```python
# delay unit = seconds
delay = 60
fileprefix = '20160928_goldbeads_t'
suffix = '.msr'
outfolder = 'C:\\Users\\RESOLFT\\Desktop\\goldbeads_over_time'


if not os.path.exists(outfolder):
    os.mkdir(outfolder)
    
im = Imspector()

interrupted = False

while not interrupted:
    try:
        start_time = time.time()
        
        print('startting measurement, current time is ' + str(time.time()))
        
        im.run(im.active_measurement())
        im.active_measurement().save_as(os.path.join(outfolder, fileprefix + str(int(time.time())) + suffix))
        
        end_time = time.time()
        
        remaining_delay = delay - (end_time - start_time)
        remaining_delay = remaining_delay if remaining_delay > 0 else 0
        
        time.sleep(remaining_delay)

    except (KeyboardInterrupt):
        interrupted = True
        continue
```
