```python
import pandas as pd
from matplotlib import pyplot as plt
import re
from Util.tile_util import *
from itertools import islice

%matplotlib inline
plt.rcParams["figure.figsize"] = [8, 8]
```

```python
dta = pd.read_csv('/Users/david/Desktop/6th_shipment_20161219/mixed_HBG2_HS2_B/GM_180sec_B_second/log.csv')

dta['x'] = dta.stg_x + dta.scan_x
dta['y'] = dta.stg_y + dta.scan_y
dta['z'] = dta.stg_z + dta.scan_z

dta['field'] = dta.file.apply(lambda x: re.match('(.*?field)([0-9]+)(.*?)', x).groups()[1])

dta_conf = dta[dta.type == 'confocal']
dta_sted = dta[dta.type == 'sted']
  
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

c = dta["type"].astype('category').cat.rename_categories([1,2])
plt.scatter(dta.x, dta.y, c=c, alpha=0.8)

#print(dta_conf.field)

z = list(zip(dta_conf.x, dta_conf.y))
for i in range(len(z)):
    ax.annotate('%s' %list(dta_conf.field)[i], xy=z[i])
    
plt.xlim(dta.x.min(), dta.x.max())
plt.ylim(dta.y.min(), dta.y.max())
plt.show()

```
