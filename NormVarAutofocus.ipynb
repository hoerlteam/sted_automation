```python
import skimage
import skimage.data
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = skimage.data.binary_blobs(100, n_dim=3)

# dimensions of image
sh = np.shape(img)

#np.var(img[:,:,3]) / np.mean(img[:,:,3])


normvar = []
for i in range(sh[2]):
    # get slice i
    tImg = img[:,:,i]
    # edge-filter slice
    tImg = skimage.filters.sobel(tImg)
    # calculate var/mean
    normvar.append(np.var(tImg) / np.mean(tImg))
    
plt.plot(normvar)

#print maximum index
print(normvar.index(max(normvar)))
```
